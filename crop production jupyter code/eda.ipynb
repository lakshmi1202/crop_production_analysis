```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\sunda\Downloads\crop_production_data.csv")
```

```python
# Step 2: Check the Data Structure
print(df.head())  # Display the first few rows of the dataset

```

```python
# Step 3: Summary Statistics
print(df.describe())  # Display summary statistics for numerical features
```

```python
# Step 4: Visualize Numerical Features
# Histograms
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# Box plots
plt.figure(figsize=(12, 8))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Density plots
plt.figure(figsize=(12, 8))
for column in df.select_dtypes(include='number').columns:
    sns.kdeplot(data=df[column], label=column)
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Density Plot of Numerical Features')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 5: Visualize Categorical Features
# Bar plots for categorical features
plt.figure(figsize=(12, 8))
for column in df.select_dtypes(include='object').columns:
    sns.countplot(x=column, data=df)
    plt.xticks(rotation=45)
    plt.title(f'Count Plot of {column}')
    plt.tight_layout()
    plt.show()
```

```python
# Step 6: Correlation Analysis
numeric_columns = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
