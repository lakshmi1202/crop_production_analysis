```python
import pandas as pd
```

```python
# Step 1: Load the Dataset
df = pd.read_csv(r"C:\Users\sunda\Downloads\crop_production_data.csv")  # Replace 'crop_production_data.csv' with your dataset file name
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
print("\nNumber of missing values in each column:")
print(df.isnull().sum())
```

```python
# Step 4: Explore Data Distributions
print("\nDescriptive statistics for numerical columns:")
print(df.describe())
```

```python
# Plot histograms for numerical columns
import matplotlib .pyplot as plt
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# Step 5: Identify Unique Values and Categories
print("\nUnique values in categorical columns:")
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].unique()}")
```

```python
# Step 6: Visualize Relationships
# Example: Scatter plot
plt.scatter(df['Area'], df['Production'])
plt.xlabel('Area')
plt.ylabel('Production')
plt.title('Production vs. Area')
plt.show()
```

```python
# Step 7: Explore Time Series Data (if applicable)
# Example: Time series plot
df['Year'] = pd.to_datetime(df['Crop_Year'], format='%Y')
plt.plot(df['Crop_Year'], df['Production'])
plt.xlabel('Crop_Year')
plt.ylabel('Production')
plt.title('Production over Time')
plt.show()
```
